# Rotate image wiht object and let it be detected

```python
import os
from PIL import Image
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
```

```python
img = Image.open("img_dogsCats.png")
```

```python
angles = range(0, 360, 1)
img_rotated_list = []
for angle in angles:
    img_rotated_list.append(img.rotate(angle))
```

```python
model = YOLO("yolov8n.pt")
# model = YOLO("yolo11n.pt")
# model = YOLO("yolo11x.pt")
# model.names
results = model.predict(source=img_rotated_list, conf=0.25, save=True)
```

```python

# Create a grid to display all results
n_images = len(results)
cols = 4
rows = (n_images + cols - 1) // cols

fig, axes = plt.subplots(rows, cols, figsize=(16, 4 * rows))
axes = axes.flatten() if n_images > 1 else [axes]

for i, result in enumerate(results):
    # Get the plotted image with detections (BGR format)
    plotted_img = result.plot()
    # Convert BGR to RGB for matplotlib
    plotted_img_rgb = cv2.cvtColor(plotted_img, cv2.COLOR_BGR2RGB)
    
    axes[i].imshow(plotted_img_rgb)
    axes[i].set_title(f"Rotation: {angles[i]}Â°")
    axes[i].axis('off')

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```
